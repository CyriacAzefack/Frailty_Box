```python
import pandas as pd
import numpy as np
```

```python
df =  pd.read_csv("data/activity_data.txt", sep=';')

df.head()
```

```python
labels_df = pd.read_csv("data/activity_labels.txt", sep=";")

labels_df['label'] = labels_df['label'].str.replace('\'', '')
labels_df.head()
```

```python
merged = pd.merge(df, labels_df, on='label_id', how='left')

merged
```

```python
new_df = pd.DataFrame(columns=["date", "activity"])

date_format = '%d-%b-%Y %H:%M:%S'
for index, row in merged.iterrows():
    new_df.loc[len(new_df)] = [row["start_time"], row["label"]+" start"]
    new_df.loc[len(new_df)] = [row["end_time"],  row["label"]+" end"]


new_df['date'] = pd.to_datetime(new_df['date'], format=date_format)
new_df.sort_values(['date'], ascending=True, inplace=True)


new_df
```

```python
new_df.to_csv("KA_dataset.csv", sep=";", index=False)
new_df.to_csv("kaData.txt", sep=";", index=False)
```
