```python
import sys
sys.path.append("../")
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
sns.set(font_scale=1.4) # for label size
```

```python
results = pd.read_csv('../output/drift_toy_delay_results_w30_l10.csv')

results['delay_days'] += 30

results
```

```python
sns.distplot(results.delay_days)
m = results.delay_days.mean()
plt.plot([m, m], [0, 0.07], linewidth=4, color='r')
plt.xlabel('Number of days')
plt.xticks(list(plt.xticks()[0]) + [m])
plt.xlim([-25, 50])
```

```python
results.nb_drifts += 1
sns.boxplot(x="nb_drifts", y="delay_days", data=results, palette="Set1")
plt.ylim([-25, 50])
plt.xlabel('Number of behaviors present')
plt.ylabel('Delay (days)')
```
