```python
import pandas as pd
import datetime as dt
import Frailty_Box.Utils
 
import locale
locale.setlocale(locale.LC_TIME,'')
```

```python
dataset = pd.read_csv('../output/hh101/habits_results.csv')

date_format = '%Y-%m-%d %H:%M:%S.%f'
dataset['start_validity'] = pd.to_datetime(dataset['start_validity'], format=date_format)
dataset['start_validity'] = dataset['start_validity'].dt.strftime("%d %b %Y")

dataset['end_validity'] = pd.to_datetime(dataset['end_validity'], format=date_format)
dataset['end_validity'] = dataset['end_validity'].dt.strftime("%d %b %Y")

dataset.sort_values(['accuracy'], ascending=False, inplace=True)

dataset


```
