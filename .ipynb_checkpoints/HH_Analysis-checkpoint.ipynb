```python
main_folder = './input/HH/'

dataset_name_list = []

for folder in os.listdir(main_folder):
    if os.path.isdir(main_folder+folder):
        dataset_name_list.append(folder)
dataset_name_list
```

```python
results_df = pd.DataFrame(columns=['dataset', 'nb_days', 'nb_labels', 'start_date', 'end_date'])


for dataset_name in dataset_name_list:
    dataset_id =  dataset_name_list.index(dataset_name)
    lvl = dataset_id * 10
    df = pick_dataset(dataset_name)
    start_date = df.date.min().to_pydatetime()
    end_date = df.end_date.max().to_pydatetime()
    
    
    duration = end_date - start_date
    nb_days = int(duration/dt.timedelta(days=1))
    
    nb_labels = len(df.label.unique())
    
    results_df.loc[len(results_df)] = [dataset_name, nb_days, nb_labels, start_date, end_date]
    
    
    # Plotting
    
    plt.text(dat.date2num(start_date), lvl, 'Dataset {}'.format(dataset_name), fontsize=16)
    ax.hlines(lvl, dat.date2num(start_date), dat.date2num(end_date), label='Dataset {}'.format(dataset_name),
                  linewidth=75, color=colors[dataset_id])
    
ax.tick_params(axis='both', which='major', labelsize=12)
fig.autofmt_xdate()
# plt.title("Activity : '{}'".format(self.label))
plt.xlabel('Time')
plt.ylabel('Datasets')
    
results_df.sort_values(['nb_days'], ascending=False, inplace=True)

results_df
    
```
