```python
import pandas as pd
import numpy as np
```

```python
df =  pd.read_csv("activity_data.txt", sep='\t')

df.head()
```

```python
labels = pd.read_csv("activity_labels.txt")

labels.index = range(1,len(labels)+1)

labels
```

```python
new_df = pd.DataFrame(columns=["date", "activity"])

for index, row in df.iterrows():
    new_df.loc[len(new_df)] = [row["start_time"], labels.loc[row["activity"], "activity"]+" start"]
    new_df.loc[len(new_df)] = [row["end_time"],  labels.loc[row["activity"], "activity"]+" end"]
    
new_df.to_csv("KA_dataset.csv", sep=";", index=False)
```
